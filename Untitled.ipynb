# SOME IMPORTANT FINDINGS TO INCREASE PROFIT

## 1.TO FIND WHICH SEGMENT AND CATEGORIES GIVES MORE PROFIT FROM TOP FIVE PROFIT MAKING STATES

```python
def state_data_viewer(states):
    product_data = df.groupby(['State'])
    for state in states :
        data = product_data.get_group(state).groupby(['Category'])
        fig, ax = plt.subplots(1 , 3, figsize = (28,5))
        fig.suptitle(state, fontsize=14)
        ax_index = 0
        for cat in ['Furniture','Office Supplies','Technology']:
            cat_data = data.get_group(cat).groupby(['Segment']).sum()
            sns.barplot(x = cat_data.Profit,y= cat_data.index, ax = ax[ax_index])
            ax[ax_index].set_ylabel(cat)
            ax_index +=1
        fig.show()
```

```python
states = ['California','Washington','New York','Michigan','Virginia']
state_data_viewer(states)
```

## 2. TO FIND MORE DEMANDABLE PRODUCTS FROM TOP FIVE STATES IN SALES AND TO CHECK DISCOUNTS IN THOSE PRODUCTS

```python
def state_data_viewer(states):
    product_data = df.groupby(['State'])
    for state in states :
        data = product_data.get_group(state).groupby(['Category'])
        fig, ax = plt.subplots(1 , 3, figsize = (28,5))
        fig.suptitle(state, fontsize=14)
        ax_index = 0
        for cat in ['Furniture','Office Supplies','Technology']:
            cat_data = data.get_group(cat).groupby(['Sub-Category']).sum()
            sns.barplot(x = cat_data.Discount,y= cat_data.index, ax = ax[ax_index])
            ax[ax_index].set_ylabel(cat)
            ax_index +=1
        fig.show()
```

```python
states = ['California','Washington','Pennsylvania','New York','Texas']
state_data_viewer(states)
```

```python
def state_data_viewer(states):
    product_data = df.groupby(['State'])
    for state in states :
        data = product_data.get_group(state).groupby(['Category'])
        fig, ax = plt.subplots(1 , 3, figsize = (28,5))
        fig.suptitle(state, fontsize=14)
        ax_index = 0
        for cat in ['Furniture','Office Supplies','Technology']:
            cat_data = data.get_group(cat).groupby(['Sub-Category']).sum()
            sns.barplot(x = cat_data.Sales,y= cat_data.index, ax = ax[ax_index])
            ax[ax_index].set_ylabel(cat)
            ax_index +=1
        fig.show()
```
